Finite-Difference acoustic seismic wave simulation
Discretization of the first-order acoustic wave equation

Compare seismograms, which are calculated with different
spatial and temporal accuracy.

This script will read in the seismograms to compare them.

##  Initialisation

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

## Read Seismograms

```python
FD_1D_DX4_DT2=np.load("Seismograms/FD_1D_DX4_DT2_fast.npy")
FD_1D_DX4_DT3_ABS=np.load("Seismograms/FD_1D_DX4_DT3_ABS_fast.npy")
FD_1D_DX4_DT4_ABS=np.load("Seismograms/FD_1D_DX4_DT4_ABS_fast.npy")
FD_1D_DX4_DT4_LW=np.load("Seismograms/FD_1D_DX4_DT4_LW_fast.npy")
FD_1D_DX8_DT2=np.load("Seismograms/FD_1D_DX8_DT2_fast.npy")
```

## Plotting

```python
plt.figure(1,figsize=(10, 6))
Line1,=plt.plot(FD_1D_DX4_DT2[0,:])
Line2,=plt.plot(FD_1D_DX4_DT3_ABS[0,:])
Line3,=plt.plot(FD_1D_DX4_DT4_ABS[0,:])
Line4,=plt.plot(FD_1D_DX4_DT4_LW[0,:])
Line5,=plt.plot(FD_1D_DX8_DT2[0,:])
plt.legend((Line1, Line2, Line3, Line4, Line5), ('FD_1D_DX4_DT2', 'FD_1D_DX4_DT3_ABS', 'FD_1D_DX4_DT4_ABS', 'FD_1D_DX4_DT4_LW', 'FD_1D_DX8_DT2'))
plt.title('Seismogram 1')
plt.ylabel('Amplitude')
plt.xlabel('Time in s')

plt.figure(2,figsize=(10, 6))
Line1,=plt.plot(FD_1D_DX4_DT2[1,:])
Line2,=plt.plot(FD_1D_DX4_DT3_ABS[1,:])
Line3,=plt.plot(FD_1D_DX4_DT4_ABS[1,:])
Line4,=plt.plot(FD_1D_DX4_DT4_LW[1,:])
Line5,=plt.plot(FD_1D_DX8_DT2[1,:])
plt.legend((Line1, Line2, Line3, Line4, Line5), ('FD_1D_DX4_DT2', 'FD_1D_DX4_DT3_ABS', 'FD_1D_DX4_DT4_ABS', 'FD_1D_DX4_DT4_LW', 'FD_1D_DX8_DT2'))
plt.title('Seismogram 2')
plt.ylabel('Amplitude')
plt.xlabel('Time in s')

plt.figure(3,figsize=(10, 6))
Line1,=plt.plot(FD_1D_DX4_DT2[2,:])
Line2,=plt.plot(FD_1D_DX4_DT3_ABS[2,:])
Line3,=plt.plot(FD_1D_DX4_DT4_ABS[2,:])
Line4,=plt.plot(FD_1D_DX4_DT4_LW[2,:])
Line5,=plt.plot(FD_1D_DX8_DT2[2,:])
plt.legend((Line1, Line2, Line3, Line4, Line5), ('FD_1D_DX4_DT2', 'FD_1D_DX4_DT3_ABS', 'FD_1D_DX4_DT4_ABS', 'FD_1D_DX4_DT4_LW', 'FD_1D_DX8_DT2'))
plt.title('Seismogram 3')
plt.ylabel('Amplitude')
plt.xlabel('Time in s');

```
